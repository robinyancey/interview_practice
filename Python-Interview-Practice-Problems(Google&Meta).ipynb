1. Two Sum (easy)

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

```python
# correct: One-pass Hash Table
from typing import List
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hash = {}
        for i in range(len(nums)):
            complement = target - nums[i]
            if complement in hash:
                return [i, hash[complement]]
            hash[nums[i]] = i
```

O(n) since hash check is only O(1) (note still O(n) if hash create doutside of the for-loop (eg. 2 for-loops is (O(2n))

Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string

```python
from typing import List
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
      ans= ''
      size = len(strs)
      strs.sort()

      maxL = min(len(strs[0]), len(strs[size-1]))

      for i in range(maxL):
        if strs[0][i] == strs[size-1][i]:
          ans += strs[0][i]
      
      return ans
```

```python
from typing import List
from typing import Optional
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        output = ListNode()
        
        # take care of edge cases
        if list1.val == None:
          output.val = list2.val
        if list2.val == None:
          output.val = list1.val


        while output.next != None:
          if list2.next > list1.next:
            output.next = list1.next
          else:
            output.next = list2.next
        return output.val
```

##swap pairs

```python
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return None
        
        has_head_changed = False
        current_node = head
        previous = None
        while current_node and current_node.next:
            next_node = current_node.next
            current_node.next = current_node.next.next
            next_node.next = current_node
            
            if previous:
                previous.next = next_node
                
            if not has_head_changed:
                head = next_node
                has_head_changed = True
            
            previous = current_node
            current_node = current_node.next
                        
        return head
```

##merge intervals

```python
intervals = [[2,6], [1,3], [8,10],[15,18]]
```

```python
class Solution:
    def merge(self, intervals: List[List[int]])-> List[List[int]]:
      intervals.sort()
      i = j = 0
      while i < len(intervals):
        j = i + 1
        while j < len(intervals):
          if intervals[i][0] < intervals[j][0]:
            if intervals[i][1] > intervals[j][0]:
              if intervals[j][1] > intervals[i][1]:
                intervals[i][1] = intervals[j][1]
                
                x = intervals[j]
                intervals.remove(x)
                j = i + 1
                continue
          
          j += 1
        i += 1

      return intervals

```

```python
def rotationalCipher(input, rotation_factor):
  # Write your code here
  

  new = ''
  for i in range(len(input)):
    
    char_alpha = ord(input[i])
    val = 0
    
    # need to separate capital and not and disclude the + 97, then % 26
    if char_alpha >= ord('A') and char_alpha <= ord('Z'):
      val = (( (ord(input[i])- ord('A')) + (rotation_factor % 26) ) % 26) + ord('A')
      
      new += chr(val)
      
    elif char_alpha >= ord('a') and char_alpha <= ord('z'):
      val = (( (ord(input[i]) - ord('a')) + (rotation_factor % 26) ) % 26) + ord('a')
      
      new += chr(val)
      
    elif char_alpha >= ord('0') and char_alpha <= ord('9'):
      val = (int(input[i]) + (rotation_factor % 10)) % 10
      
      new += str(val)
    else:
      new += input[i]
      
      
  return new
```

```python
# Write any import statements here
from math import floor, log10
def getUniformIntegerCountInInterval(A: int, B: int) -> int:
  def ndigits(num): # in order to make soln O(1) by not converting to string taking O(log(n)) time
    digits = floor(log10(num)) + 1
    return digits 
  
  tot = 0 # store the resulting count of uniform #s
   #  we also need to subtract off double counted values if length is same
  tot += (ndigits(B) - ndigits(A) - 1) * 9 
    # accounts for 9 different 10s multiple uniform numbers in between 
    # (eg. len('400')  - len('2') - 1 = 1 since we account for 11,22,33, etc.
  # obviously the above skips the numbers with the same length for each 
  # so now lets count how many will be included based on how large the other digits are compared to the first one
  # eg for the lower number we can count everything above and the higher number everything below (in the same 10s range)
  # first convert to int and list to compare number values separately
  # since A is lower we subtract the first digit from 9 to get values above that will  be included in the range B-A
  # eg. so if its '80' we include 99 (since 9-8=1)...
  maxA = 10**ndigits(A)
  maxB = 10**ndigits(B)

  tot += 9 - floor(A/(maxA / 10))#int(str(A)[0])

  # then we need everything up to the higher number (eg. if its '300' we need 111, 222, and maybe 333 (depending on the other digits being lower)...
  tot += floor(B/(maxB / 10))
  # but we also would need 88 in the case that  the rest of the digits are same (since interval is inclusive) or lower 
  floorA = floor((1/9)*(maxA))
  floorB = floor((1/9)*(maxB))

  test = floorA * floor(A/(maxA / 10))
  if A <= test:
    tot += 1
    # so just subtract 1 if other digits are lower and its not just returning 0
    # since all digits are the same
  test = floorB * floor(B/(maxB / 10))

  if B < test:
    tot -= 1

  return tot

print(getUniformIntegerCountInInterval(75, 300))
print('expected: 5')
print(getUniformIntegerCountInInterval(1,9))
print('expected: 9')

print(getUniformIntegerCountInInterval(9999999,9999999))
print('expected: 1')
print(getUniformIntegerCountInInterval(777,887))
print('expected: 1')
```

```python
# Write any import statements here

def getArtisticPhotographCount(N: int, C: str, X: int, Y: int) -> int:
    count = 0
    leftP, rightP, leftB, rightB = 0, 0, 0, 0
    # reduce time complexity by counting B and P in a sliding window between X and Y on each side
    if C[0] == "B": 
        leftB = 1
    elif C[0] == "P": 
        leftP = 1
    for i in range(2*X, min(N, X+Y+1)): 
        # max upper bound N, lower right starts i >= 2X
        if C[i] == "B": rightB += 1
        if C[i] == "P": rightP += 1
    if X != N and C[X] == "A": 
        # means no interval since range with Y adds no distance
        count = leftB*rightP + rightB*leftP
    for i in range(X+1,N-X): # rest of range where A could count  
        if i>Y and C[i-Y-1] == "B": # remove one left B since we accounted for it above and are now moving out of the range
            leftB -= 1
        if i>Y and C[i-Y-1] == "P": 
            leftP -= 1
        # as we move right we can move things into left side interval if they are X away at least
        if C[i-X] == "B": 
            leftB += 1
        if C[i-X] == "P": 
            leftP += 1
        # things can also move into the right window (between X and Y) incrementally
        if i+Y<N and C[i+Y] == "B": 
            rightB += 1
        if i+Y<N and C[i+Y] == "P": 
            rightP += 1
        # or move out of it
        if C[i+X-1] == "B": rightB -= 1
        if C[i+X-1] == "P": rightP -= 1
        if C[i] == "A": count += leftB*rightP + rightB*leftP
    return count


print(getArtisticPhotographCount(5, 'APABA', 1, 2), 'expected: 1')
print(getArtisticPhotographCount(5, 'APABA', 2, 3), 'expected: 0')
print(getArtisticPhotographCount(8, '.PBAAP.B', 1, 3), 'expected: 3')
print(getArtisticPhotographCount(9, 'PBBAAPP.B', 1, 6), 'expected: 10')
```

```python
from typing import List
def bs(array: List[int], x:int): 
    array.sort()
    med = (len(array) -1) // 2
    Left = 0
    Right = len(array)-1
    while Left <= Right:
        if x < array[med]:
            med = Left - 1
        elif x > array[med]:
            med = Right + 1
        elif x == array[med]:
            return med, array
    return ValueError(x,'not in the list')

print(bs([7,8,9,1,2,3,4], 4))
```

```python
class MyStack:

    def __init__(self):
        self.qstack = []

    def push(self, x: int) -> None:
        self.qstack.append(x)
        
    def rev(self):
        self.temp = []
        for i in range(len(self.qstack)):
            x = self.qstack[-1]
            self.temp.append(x)
        
        self.qstack = self.temp

    def pop(self) -> int:
        self.rev()
        x = self.qstack.pop(0)
        self.rev()
        return x 

    def top(self) -> int:
        self.rev()
        x = self.qstack[0]
        self.rev()
        return x 

    def empty(self) -> bool:
        return len(self.qstack) == 0



myStack = MyStack()
print(myStack.push(1))
print(myStack.push(2))
print(myStack.top())# should return 2
print(myStack.pop())# shouldreturn 2
print(myStack.empty())# should return False
```

##q using stacks

```python
class MyQueue:

    def __init__(self):
        self.stackq = []

    def push(self, x: int) -> None:
        self.stackq.append(x)
    def rev(self):
        self.temp = []
        for i in range(len(self.stackq)-1,-1,-1):
            self.temp.append(self.stackq[i])
        self.stackq = self.temp
        
    def pop(self) -> int:
        self.rev()
        save = self.stackq[-1]
        del self.stackq[-1]
        self.rev()
        return save

    def peek(self) -> int:
        self.rev()
        save = self.stackq[-1]
        self.rev()
        return save     

    def empty(self) -> bool:
       
        if (len(self.stackq) > 0):
            return False
        else:
            return True
        


myQueue = MyQueue()
myQueue.push(1)#queue is: [1]
myQueue.push(2)# queue is: [1, 2] (leftmost is front of the queue)
print(myQueue.peek())# return 1
print(myQueue.pop())# return 1, queue is [2]
print(myQueue.empty())# return false
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def isPalindrome(head: ListNode) -> bool:
    temp = head
    save = []
    while temp is not None:
      save.append(temp.val)
      temp = temp.next

    for i in reversed(save):
        if head.val != i:
            return False
        head = head.next

    return True



def createLL(list):
  temp = None
  for i in reversed(list):
      temp = ListNode(i,temp)

  return temp


print(isPalindrome(createLL([1,2])), 'expected: false')
print(isPalindrome(createLL([1,2,2,1])), 'expected: true')
```

```python
from typing import List
import unittest

class soln:
    def f(self, board: List[List[str]], word: str)-> bool:
      i = j = k = 0
      start_i, start_j = 0, 0 # to keep track of our current start loc
      path = []
      while k < len(word):
        # if we are somewhere on the path of the word, one of the 
        # follwoing will be true
        if k == 0 and board[i][j] == word[k]:
            k += 1
            path.append([i,j])
            
        elif (k > 0 and j < len(board[0])-1 and board[i][j+1] == word[k] and [i,j+1] not in path):
            path.append([i,j+1])
            k += 1; j += 1            
            
        elif (k > 0 and j > 0 and board[i][j-1] == word[k] and [i,j-1] not in path):
            path.append([i,j-1])
            k += 1; j -= 1            
            
        elif (k > 0 and i < len(board)-1 and board[i+1][j] == word[k] and [i+1,j] not in path):
            path.append([i+1,j])
            k += 1; i +=1           
            
        elif (k > 0 and 0 < i and board[i-1][j] == word[k] and [i-1,j] not in path):
            path.append([i-1,j])
            k += 1; i -=1
            
           
        else:
            # start over word from new spot
            k = 0; path = []
            if start_j < len(board[0])-1:
                start_j += 1
                j =  start_j; i = start_i
            elif start_i < len(board)-1:
                start_i += 1; start_j= 0
                j =  start_j; i = start_i
            else:
                # else we cant make the word
                return False
        
      return True



class test_LC(unittest.TestCase):
  def test_f(self):
      test = soln()
      self.assertEqual(test.f(board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "ABCCED"),True)
      self.assertEqual(test.f(board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "SEE"),True)
      self.assertEqual(test.f(board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]], word = "ABCB"),False)

if __name__ == '__main__':
  unittest.main(argv=['first-arg-is-ignored'], exit=False)

```

```python
from typing import List
import unittest

class soln:
    def f(self, matrix: List[List[int]]):
        row = 0
        while row < (len(matrix) // 2): # ignore any center in odd size
          for i in range(row,len(matrix)-row-1):
            # can do 3 swaps
              # 1. swap with one in the right side (ith element going down the last column)
            matrix[row][i], matrix[i][len(matrix)-row-1] = matrix[i][len(matrix)-row-1], matrix[row][i]
              # 2. swap with one on the bottom (decreasing by i in the last row)
            matrix[row][i], matrix[len(matrix)-row-1][len(matrix)-1-i] = matrix[len(matrix)-row-1][len(matrix)-1-i], matrix[row][i]
              # 3. swap with one on the left side (going up i in the first column)
            matrix[row][i], matrix[len(matrix)-1-i][row] = matrix[len(matrix)-1-i][row], matrix[row][i]
          row += 1
        return matrix
      



class test_LC(unittest.TestCase):
  def test_f(self):
      test = soln()
      self.assertEqual(test.f(matrix = [[1,2,3],[4,5,6],[7,8,9]]),[[7,4,1],[8,5,2],[9,6,3]])
      self.assertEqual(test.f(matrix = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]),[[15,13,2,5],[14,3,4,1],[12,6,8,9],[16,7,10,11]])


if __name__ == '__main__':
  unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

## LRU Cache

```python
class LRUCache:

    def __init__(self, capacity: int):
        self.cache = {} # just holds the values
        self.cap = capacity # max size of cache
        self.LRU = [] #queue

        

    def get(self, key: int) -> int:
        if key not in self.cache:
            return -1
        else:
            self.LRU.pop(self.LRU.index(key))
            self.LRU.append(key)
            return self.cache[key]
   

    def put(self, key: int, value: int) -> None:

        if key not in self.cache and len(self.LRU) == self.cap:
            rem = self.LRU.pop(0)
            del self.cache[rem]

        elif key in self.cache:            
            self.LRU.pop(self.LRU.index(key))
        self.LRU.append(key)
        self.cache[key] = value

        return None

        

def f(c,inp):
  res = []
  for i in range(len(c)):
    if c[i] == 'LRUCache':
        temp = LRUCache(inp[i][0])
        res.append(None)
    elif c[i] == 'put':
        res.append(temp.put(inp[i][0],inp[i][1]))
    elif c[i] == 'get':
        res.append(temp.get(inp[i][0]))
  
  return res
    
   

import unittest

class test_LC(unittest.TestCase):
  def test_f(self):
    self.assertEqual( f(["LRUCache", "put", "put", "get", "put", "get", "put", "get", "get", "get"],[[2], [1, 1], [2, 2], [1], [3, 3], [2], [4, 4], [1], [3], [4]]),[None, None, None, 1, None, -1, None, -1, 3, 4])


if __name__ == '__main__':

  unittest.main(argv=['first-arg-is-ignored'], exit=False)
```

```python
from typing import List
class Node:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class LL:
    def __init__(self, head: None):
        self.head = Node(head)


class Solution:
    #def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    def createLL(self, ll: List[int]):
        res = LL(ll[0])
        
        i = 1
        temp = res.head
        while i < len(ll):
            temp.next = Node(ll[i])
            temp = temp.next
            i +=1 
        
        return res.head

    def printLL(self, head: Node):
      temp = head
      print('Structure of LL:')
      while temp:
        print('Node val:', temp.val, end=', ')
        if temp.next == None:
            print('with next: None', end=' ')
        else:
            print('with next', temp.next.val, end=' ')
        print('  ')
        temp = temp.next
      return        

    def split(self, head):
        if head == None or head.next == None:
            return (head, None)
        
        fast = head.next
        slow = head

        while fast:
            fast = fast.next
            if fast:
                slow = slow.next
                fast = fast.next

        res = (head, slow.next)
        slow.next = None

        return res

    def connect(self, start, end):
        if start == None:
            return end
        elif end == None:
            return start

        if start.val <= end.val:
          res = start
          res.next = self.connect(start.next, end)
        else:
          res = end
          res.next = self.connect(end.next, start)

        return res

    def ms(self, head):
        if head == None or head.next == None:
            return head

        start, end = self.split(head)

        start = self.ms(start)
        end = self.ms(end)

        return self.connect(start, end)

        

if __name__ == '__main__':
    S = Solution()
    res1 = S.createLL([4,2,1,3])
    
    print('Original', end= ' ')
    S.printLL(res1)
    res2 = S.ms(res1)
    print('Sorted', end= ' ')
    S.printLL(res2)
```

## modified closest k elements

```python
def f(nums, k, x):

    ind = nums.index(x)
    i = ind - 1
    j = ind + 1

    if ind == 0:
        return nums[1:5]
    elif ind == len(nums)-1:
        return nums[-5:-1]

    if len(nums) <= k+1:
        return nums[0:ind] + nums[ind+1:]
    res = []

    while len(res) < k and (i >= 0 or j <= len(nums)-1):

        if i < 0:
            res.append(nums[j])
            j += 1
        elif j > len(nums) - 1:
            res.append(nums[i])
            i -= 1
        elif abs(nums[i]-nums[ind]) < abs(nums[j]-nums[ind]):
              res.append(nums[i])
              i -= 1
        else:
              res.append(nums[j])
              j +=1

    return sorted(res)



print('Actual:',f([1,2,3,4,5], k = 4, x = 5), 'expected:', [1,2,3,4])
print('Actual:',f([1,2,3,4,5], k = 4, x = 3), 'expected:', [1,2,4,5])
print('Actual:',f([1,2,3,3,7,7,7,9,20], k = 4, x = 3), 'expected:', [1,2,3,7])
```
